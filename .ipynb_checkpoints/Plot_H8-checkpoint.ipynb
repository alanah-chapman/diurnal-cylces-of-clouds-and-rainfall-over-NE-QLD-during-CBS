# interpolate and plot for single time

```python
# if you just want to look at a single timestamp, specify path
fp = "/g/data/v46/lb5963/HIMAWARI/08_V46_SUMM_2021-22/L1/geocatL1.HIMAWARI-8.2022025.050000.FLDK.R20.nc"
```

```python
# open H8 file
ds = xr.open_dataset(fp,engine="netcdf4") 

# variable to interpolate - for valid points (i used the iris package to visualize these and for creating smaller domains
# to decrease compute time when looking at multiple files)
ch13bt_data = ds['himawari_8_ahi_channel_13_brightness_temperature'][:3299,33:4124].values  
ch13bt_reshape = ch13bt_data.reshape(13496209)  # scipy.interpolate griddata needs 1D array inputs
ch13bt_reshape[np.isnan(ch13bt_reshape)] = 0    # set any nan values to 0
# longitude values for matching data to np.meshgrid
lon2d = ds['pixel_longitude'][:3299,33:4124].values
lon1d = lon2d.reshape(13496209)
lon1d[np.isnan(lon1d)] = 0
# latitude values
lat2d = ds['pixel_latitude'][:3299,33:4124].values
lat1d = lat2d.reshape(13496209)
lat1d[np.isnan(lat1d)] = 0
    
# create meshgrid for interpolation
res = 0.1 # resolution of grid
# create arrays for domain of interest
grid_x = np.arange(143, 154, res)
grid_y = np.arange(-22, -11, res)
# so y is right direction
grid_y = np.flipud(grid_y)
# create grid
xi,yi = np.meshgrid(grid_x,grid_y)
# interpolate channel 13 brightness temperature values onto x,y meshgrid
BT_data = griddata((lon1d,lat1d), ch13bt_reshape, (xi, yi), method='nearest') ##### returns= BT_data[lat:lon] 
```

```python
#### create snapshot latlong plot
ax = plt.axes(projection=ccrs.PlateCarree())
levels = [180,190,200,210,220,230,240,250,260,270,280,290,300,310,320]
cs = ax.contourf(xi,yi,BT_data,levels=levels,cmap="binary")
cbar = plt.colorbar(cs,aspect=20,orientation='vertical',ax=ax,
                    ticks=[180,190,200,210,220,230,240,250,260,270,280,290,300,310,320])
cbar.set_label('K',fontsize=20) 
gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, linewidth=0.5, color='gray', alpha=0.5, linestyle='-')
gl.top_labels = False
gl.right_labels = False
ax.set_xlim(143,152)
ax.set_ylim(-22,-15)
ax.coastlines(color="white")
```

# interpolate H8 data for multiple file timeseries

```python
# path for files - for multiple:
mypath = "/g/data/v46/lb5963/HIMAWARI/08_V46_SUMM_2021-22/L1/"
files = [os.path.join(root, f) for root, _, files in os.walk(mypath)
         for f in files
         if f.startswith('geocatL1.HIMAWARI-8.2022') and f.endswith('.nc')]
files.sort()
```

```python
# create empty arrays to append data
BT_data_T = []
BT_data_C = []
BT_data_W = []
data = []
times = []

# create for loop to generate data arrays
for file in files[0:3]:   
  # read file 
    ds = xr.open_dataset(file,engine="netcdf4") 
    
  #assign file variable contents to python variables and reshape - making smaller domain decreased compute time significantly
        # used iris cubes (below) to visualize data in order to choose domain size
    ch13bt_data = ds['himawari_8_ahi_channel_13_brightness_temperature'][:3299,33:4124].values  #for valid points
    ch13bt_reshape = np.reshape(ch13bt_data,(13496209)) 
    ch13bt_reshape[np.isnan(ch13bt_reshape)] = 0    
    lon2d = ds['pixel_longitude'][:3299,33:4124].values
    lon1d = lon2d.reshape(13496209)
    lon1d[np.isnan(lon1d)] = 0
    lat2d = ds['pixel_latitude'][:3299,33:4124].values
    lat1d = lat2d.reshape(13496209)
    lat1d[np.isnan(lat1d)] = 0
    
  # create meshgrid for interpolation
    res = 0.1 # resolution of grid
  # create arrays for domain of interest
    grid_x = np.arange(143, 154, res)
    grid_y = np.arange(-22, -11, res)
  # so y is right direction
    grid_y = np.flipud(grid_y)
  # create grid
    xi,yi = np.meshgrid(grid_x,grid_y)

  # interpolate channel 13 brightness temperature values onto x,y meshgrid
    BT_data = griddata((lon1d,lat1d), ch13bt_reshape, (xi, yi), method='nearest') ##### returns= BT_data[lat:lon] 
  # i am looking at data for specified latitudes
    BT_data_T.append(BT_data[83,:]) # townsville latitude index for my meshgrid (checked by looking at yi values)
    BT_data_C.append(BT_data[57,:]) 
    BT_data_W.append(BT_data[52,:]) 
  # if you want all values interpolated onto the domain of the meshgrid 
    data.append(BT_data[:,:])
    
  # append time data (from file name) for use in plotting
    fn = os.path.basename(file).split('/')[-1]
    times.append(fn[20:32])
    
```
