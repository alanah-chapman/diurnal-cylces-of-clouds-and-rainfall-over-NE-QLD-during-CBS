# This notebook contains:
* Code to create a gif file from images saved in a particular file

```python
import glob
from PIL import Image
# filepaths
fp_in = "/home/563/ac9768/plots/case_study/*.png"
fp_out = "/home/563/ac9768/plots/case_study-jan.gif"

# https://pillow.readthedocs.io/en/stable/handbook/image-file-formats.html#gif
img, *imgs = [Image.open(f) for f in sorted(glob.glob(fp_in))]
img.save(fp=fp_out, format='GIF', append_images=imgs,save_all=True, duration=150, loop=0)
```
